## Анализ корреляции признаков, построение heatmap

```python
X_corrmat=X.corr()
X_corrmat.loc[:,:] =  np.tril(X_corrmat, k=-1) # borrowed from Karl D's answer
already_in = set()
result = []
for col in X_corrmat:
    perfect_corr = X_corrmat[col][abs(X_corrmat[col]) > 0.5].index.tolist()
    if perfect_corr and col not in already_in:
        already_in.update(set(perfect_corr))
        perfect_corr.append(col)
        result.append(perfect_corr)
```

```python
del_cols=list()
for r in result:
    for f in r[1:]:
        del_cols.append(f)
X_del=X.drop(del_cols,axis=1)
```

```python
X_corrmat1=X_del.corr()
```

```python
sns.set(context="paper", font="monospace")
plt.figure(figsize=(50, 50))
sns_heatmap = sns.heatmap(X_corrmat1, vmax=.8,annot=True, fmt=".2f", linewidths=.5)
sns_heatmap.get_figure().savefig("heatmap1.png")
#plt.show()
```

## Попарное PCA

```python
from sklearn.decomposition import RandomizedPCA,PCA
from sklearn import manifold
from matplotlib.colors import ListedColormap
from sklearn import random_projection
from sklearn.model_selection import cross_val_score as cv_score
colors = ['red','green','blue','purple','yellow']
def plot_scores(d_scores,d_opt):
    n_components = np.arange(1,d_scores.size+1)
    plt.plot(n_components, d_scores, 'b', label='PCA scores')
    plt.xlim(n_components[0], n_components[-1])
    plt.xlabel('n components')
    plt.ylabel('cv scores')
    plt.legend(loc='lower right')
    plt.axvline(d_opt,ls='dotted')
    plt.show()
```

```python
pca = PCA(n_components=2,random_state=2,svd_solver='full')
data_2d_pca = pca.fit_transform(X)
plt.figure(figsize = (10, 6))
plt.scatter(data_2d_pca[:, 0], data_2d_pca[:, 1], c = y,cmap=ListedColormap(colors))
plt.show()
```

```python
pca = PCA(n_components=2,random_state=2,svd_solver='full')
k=0
plt.figure(figsize = (30, 30))
for i in range(5):
    for j in range(i+1,5):
        k=k+1
        class_filter=(y['class']==i)|(y['class']==j)
        X_cl=X[class_filter] 
        y_cl=y[class_filter]   
        data_2d_pca = pca.fit_transform(X_cl)
        plt.subplot(5,2,k)
        plt.scatter(data_2d_pca[:, 0], data_2d_pca[:, 1], c = y_cl,cmap=ListedColormap(colors))
plt.show()
```
